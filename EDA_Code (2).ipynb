```python
import pandas as pd
```

```python
df = pd.read_excel('C:\\Users\\nikhi\\OneDrive\\Desktop\\Bhoomika\\Data Analytics\\Project\\Data_Set_EDA.xlsx')
#print(df)
df2=df.copy()
```

```python
list_of_yesorno_questions = ['Do you work well under pressure?', 'Do you think defining goals important before embarking on a project/team task?', 'Are you willing to take responsibility when a team member fails to deliver against expectations?', 'Do you like to act on feedback and suggestions from your team?', 'Are you open to criticism and constructive feedback from team members.', 'Would you describe yourself as a team player?', 'Have you held a leadership position before?']
for i in list_of_yesorno_questions:
    df2[i] = df[i].replace({'Yes': 1, 'No': 0})
    #print(df2[i])
```

```python
list_of_nottruetoverytrue_questions = ['I am effective with the detailed aspects of my work', 'Managing people and resources is one of my strengths', 'I usually know ahead of time how people will respond to a new idea or proposal', 'Understanding the social fabric of the organization/team is important to me', 'I am able to sense the emotional undercurrents in my team.', 'The key to successful conflict resolution is respecting my opponent', 'I work hard to find consensus in conflict situations', 'I am effective at problem solving', 'When problems arise, I immediately address them', 'Seeing the big picture comes easily for me( Issues /Concern)', 'Making strategic plans for my company/team appeals to me', 'I enjoy discussing values and philosophy(Organizational/Team)', 'I am flexible about making changes in our organization/team', 'I encourage people to persevere during difficult times', 'I involve people at all levels of the organization in implementing a change initiative', 'I must not hold any grudges or biases against anyone in the team.', 'I am open to suggestions from employees and co-workers/team members', 'I try to understand the issues of the team', 'I have the ability to sense other people\'s emotions, coupled with the ability to imagine what someone else might be thinking', 'I am good at brainstorming', 'I would feel comfortable asking team members opinions and ideas for decision-making', 'I have good sense of which team has the greatest potentials terms of helping the organization to achieve its long term goals/objectives', 'I have a clear vision of where my team is headed.', 'I must behave in a trustworthy way to set an example for my team members.', 'I state clearly the goals that team members should be working towards']
for i in list_of_nottruetoverytrue_questions:
    #df2 = df[i]
    df2[i]=df[i].replace({'Not true': 0.2, 'Seldom true': 0.4, 'Occasionally true':0.6, 'Somewhat true':0.8, 'Very true': 1}, regex=True)
    #print(df2[i])
```

```python
#df2 = df["How well do you delegate tasks to other team members?"]
df2["How well do you delegate tasks to other team members?"]=df["How well do you delegate tasks to other team members?"].replace({'Very well': 0.99, 'Somewhat well': 0.66, 'Not at all':0.33}, regex=True)
#print(df2["How well do you delegate tasks to other team members?"])
#df2 = df["On a scale of 1–5, how well you can collaborate with others?"]
df2["On a scale of 1–5, how well you can collaborate with others?"]=df["On a scale of 1–5, how well you can collaborate with others?"].replace({'1': 0.2, '2': 0.4, '3': 0.6, '4':0.8, '5': 1}, regex=True)
#print(df2)
#df2 = df["How well do you interact with team members?"]
df2["How well do you interact with team members?"]=df["How well do you interact with team members?"].replace({'Very Large Extent': 1, 'Large Extent': 0.8, 'Neutral': 0.6, 'Little Extent': 0.4, 'Very Little Extent': 0.2}, regex=True)
#print(df2)
#df2 = df["I motivate my team members to do their best on the job/task."]
df2["I motivate my team members to do their best on the job/task."]=df["I motivate my team members to do their best on the job/task."].replace({'Always or almost always': 1, 'Frequently or often': 0.75, 'Sometimes or reasonably': 0.5, 'Rarely or never': 0.25},regex=True)
#print(df2)
#df2 = df["Do you feel providing guidance without any pressure is a trait of a good leader?"]
df2["Do you feel providing guidance without any pressure is a trait of a good leader?"]=df["Do you feel providing guidance without any pressure is a trait of a good leader?"].replace({'Strongly Disagree': 0.2, 'Disagree': 0.4, 'Neutral': 0.6, 'Agree': 0.8, 'Strongly agree': 1 }, regex=True)
#print(df2)
#df2 = df["Obtaining and allocating resources is a challenging aspect of my job"]
df2["Obtaining and allocating resources is a challenging aspect of my job"]=df["Obtaining and allocating resources is a challenging aspect of my job"].replace({'Not true': 1, 'Seldom true': 0.8, 'Occasionally true': 0.6, 'Somewhat true': 0.4, 'Very true': 0.2 }, regex=True)
#print(df2)
#df2 = df["I use my emotional energy to motivate others."
df2["I use my emotional energy to motivate others."]=df["I use my emotional energy to motivate others."].replace({'Not true': 0.33, 'Seldom true': 0.66, 'Occasionally true': 0.99, 'Somewhat true': 0.66, 'Very true': 0.33 }, regex=True)
#print(df2)
#df2 = df["I would hesitate to ask advice from team members"]
df2["I would hesitate to ask advice from team members"]=df["I would hesitate to ask advice from team members"].replace({'Not true': 1, 'Seldom true': 0.8, 'Occasionally true': 0.6, 'Somewhat true': 0.4, 'Very true': 0.2 }, regex=True)
#print(df2)
#df2 = df["Do you feel uncomfortable when someone questions your decisions?"
df2["Do you feel uncomfortable when someone questions your decisions?"]=df["Do you feel uncomfortable when someone questions your decisions?"].replace({'Not at all': 1, 'Rarely': 0.8, 'Sometimes': 0.6, 'Often': 0.4, 'Very Often': 0.2 }, regex=True)
#print(df2)
#df2 = df["Employees must carry out orders, no questions asked"]
df2["Employees must carry out orders, no questions asked"]=df["Employees must carry out orders, no questions asked"].replace({'Strongly Disagree': 1, 'Disagree': 0.8, 'Neutral': 0.6, 'Agree': 0.4, 'Strongly Agree': 0.2 }, regex=True)
#print(df2)
```
